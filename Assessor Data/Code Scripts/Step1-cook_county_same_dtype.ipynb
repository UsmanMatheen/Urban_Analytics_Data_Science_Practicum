## Data Type Standardization for Cook County Assessor Dataset

This notebook processes the Cook County Assessor’s “Single and Multi-Family Improvement Characteristics” dataset to ensure data consistency by standardizing column data types.

### Objective:
The original dataset contained several columns with **mixed data types** — for example, some columns had both integers and strings representing the same kind of information. This inconsistency can cause issues in analysis, filtering, joins, and database storage.

### What We Did:
- Identified columns with **mixed data types** using Python’s `type()` function and `.map()`.
- Converted **all values** within such columns to a **uniform type**, mostly `string`, to ensure consistency.
- Processed the dataset in **chunks** (due to its size) for memory efficiency.
- Verified that all columns now contain a **single, consistent data type**.

This cleaning process is crucial for ensuring reliable downstream analysis, modeling, and database ingestion.

✅ Final outputs contain columns with unified data types, improving data quality and usability.

```python
import numpy as np

# File paths
input_file = "processed_cook_county_data.csv"
output_file = "final_cleaned_cook_county_data.csv"

# Define a safe chunk size (adjust if needed)
chunk_size = 50000  

# Define the expected data types
dtype_mapping = {
    "pin": "str",
    "tax_year": "int64",
    "card_num": "int64",
    "class": "str",
    "township_code": "str",
    "proration_key_pin": "str",
    "pin_proration_rate": "int64",
    "card_proration_rate": "int64",
    "cdu": "str",
    "pin_is_multicard": "category",
    "pin_num_cards": "int64",
    "pin_is_multiland": "category",
    "pin_num_landlines": "int64",
    "year_built": "int64",
    "building_sqft": "int64",
    "land_sqft": "int64",
    "num_bedrooms": "int64",
    "num_rooms": "int64",
    "num_full_baths": "int64",
    "num_half_baths": "int64",
    "num_fireplaces": "int64",
    "type_of_residence": "str",
    "construction_quality": "str",
    "num_apartments": "int64",
    "attic_finish": "str",
    "garage_attached": "str",
    "garage_area_included": "str",
    "garage_size": "str",
    "garage_ext_wall_material": "str",
    "attic_type": "str",
    "basement_type": "str",
    "ext_wall_material": "str",
    "central_heating": "str",
    "repair_condition": "str",
    "basement_finish": "str",
    "roof_material": "str",
    "single_v_multi_family": "str",
    "site_desirability": "str",
    "num_commercial_units": "int64",
    "renovation": "str",
    "porch": "str",
    "central_air": "str",
    "design_plan": "str"
}

# Flag to write header only for the first chunk
first_chunk = True

# Process file chunk by chunk and convert dtypes while saving
for chunk in pd.read_csv(input_file, chunksize=chunk_size):

    # Fix NaN and non-numeric values before converting to int
    for col, dtype in dtype_mapping.items():
        if dtype == "int64":
            chunk[col] = pd.to_numeric(chunk[col], errors="coerce")  # Convert invalid strings to NaN
            chunk[col] = chunk[col].fillna(-1).astype("int64")  # Replace NaN with -1 and convert to int

        else:
            chunk[col] = chunk[col].astype(dtype)  # Convert non-int columns directly

    # Append to CSV (write header only for the first chunk)
    chunk.to_csv(output_file, mode="a", index=False, header=first_chunk)
    first_chunk = False  # Ensure header is written only once

print("✅ Final cleaned dataset saved as 'final_cleaned_cook_county_data.csv' successfully!")
```

```python
# Update the file name based on what you find
output_file = "final_class_int_quoted_only.csv"  # Ensure this file exists

# Dictionary to store value counts
class_value_counts = {}

# Read dataset in chunks and count occurrences of each unique value
chunk_size = 50000  # Adjust based on memory capacity

for chunk in pd.read_csv(output_file, usecols=["class"], chunksize=chunk_size):
    chunk_counts = chunk["class"].value_counts().to_dict()
    
    # Aggregate counts across chunks
    for key, value in chunk_counts.items():
        class_value_counts[key] = class_value_counts.get(key, 0) + value

# Convert to DataFrame for better visualization
class_value_counts_df = pd.DataFrame(
    sorted(class_value_counts.items(), key=lambda x: x[1], reverse=True), 
    columns=["class_value", "count"]
)

# Print the results
print("\n📊 Unique Values in 'class' Column and Their Counts:\n")
print(class_value_counts_df)
```
