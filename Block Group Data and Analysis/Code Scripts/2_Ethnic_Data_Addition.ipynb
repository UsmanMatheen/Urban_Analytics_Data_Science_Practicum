```python
import pandas as pd
```

# Part 1

```python
df = pd.read_csv(r"E:\IIT Chicago\Sem 4\Data Science Practicum - CSP-572\Analysis\Dataset Files\Ethnic Data\ACSDT5Y2023.B02001-Data.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
df = df.drop(df.index[0]).reset_index(drop=True)
```

```python
df.info()
```

```python
df.columns = df.columns.astype(str)
```

```python
df = df.loc[:, ~df.columns.str.contains('Margin')]
```

```python
df.info()
```

```python
df.drop('nan', axis = 1, inplace = True)
```

```python
df.columns = df.columns.str.replace('^Estimate!!', '', regex=True)
```

```python
df.info()
```

```python
df.columns = df.columns.str.replace('!!', '_', regex=False).str.replace(':', '', regex=False)
```

```python
df.info()
```

```python
df.columns = [
    col if col == 'Total' else col.replace('Total_', '')
    for col in df.columns
]
```

```python
df.info()
```

```python
df.drop('Geographic Area Name', axis = 1, inplace  = True)
```

```python
df.head()
```

```python
df['geoid'] = df['Geography'].str[-12:]
```

```python
df.drop('Geography', axis = 1, inplace  = True)
```

```python
df.info()
```

```python
df.isnull().sum()
```
